```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('AQI.csv')
df.head()
```

```python
df.tail()
```

```python
df.info()
```

```python
#checking if there are null values in dataset
df.isnull().sum()
```

```python
#removing row where date is null
df['Date'] = pd.to_datetime(df['Date'])
df.dropna(subset=['Date'], inplace = True)
```

```python
df.info()
```

```python
df.head()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.shape
```

```python
df.columns
```

```python
# df.plot()

plt.bar(df['Country'],df['AQI Value'])
plt.xlabel('country')
plt.ylabel('AQI value')
plt.xticks(rotation='vertical')
# plt.show()
```

```python
df.Country.unique()
```

```python

df = df.set_index("Date")
```

```python
#top 20 countries with high AQi
df.nlargest(20,'AQI Value')
```

```python
df.nsmallest(20,'AQI Value')
```

```python
major_countries = df[df['Country'].isin(['India', 'USA','China','Brazil','France','Russia','Great Britain','Japan','Germany','South Africa']) ].reset_index()
major_countries
```

```python
haqi = major_countries.nlargest(20,'AQI Value')
haqi
```

```python
laqi= major_countries.nsmallest(20,'AQI Value')
laqi
```

```python
#bar plot of countries with high aqi
plt.bar(haqi['Country'],haqi['AQI Value'])
plt.xlabel('country')
plt.ylabel('AQI value')
plt.title('bar plot of countries with high aqi')
plt.xticks(rotation='vertical')
plt.show()
```

```python
#bar plot of countries with low aqi
plt.bar(laqi['Country'],laqi['AQI Value'])
plt.xlabel('country')
plt.ylabel('AQI value')
plt.title('bar plot of countries with low aqi')
plt.xticks(rotation='vertical')
plt.show()
```

```python
#Scatterplot of high aqi country
plt.scatter(haqi['Country'], haqi['AQI Value'], marker='o')
plt.xlabel('country')
plt.ylabel('AQI value')
plt.title('Scatterplot of high aqi country')
plt.show()
```

```python
#seaborn
sns.relplot(data=haqi,x='Country',y='AQI Value',kind='scatter',hue='AQI Value')
plt.title('Scatterplot of high aqi country')
```

```python
#Scatterplot of low aqi country
plt.scatter(laqi['Country'], laqi['AQI Value'], marker='o')
plt.xlabel('country')
plt.ylabel('AQI value')
plt.title('Scatterplot of low aqi country')
plt.show()
```

```python
#seaborn
sns.relplot(data=laqi,x='Country',y='AQI Value',kind='scatter',hue='AQI Value')
plt.title('Scatterplot of low aqi country')
```
